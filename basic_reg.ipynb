### Basic regression case example

```python
# Import libraries
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Create basic data
X = torch.randn(200, 1)
y = 3*X - 2 + 0.1 * torch.randn(200, 1)
```

```python
# Basic neural network
model = nn.Sequential(
    nn.Linear(1, 16), # I tried with other values replacing 16
    nn.ReLU(),
    nn.Linear(16, 1)
)
```

```python
# Loss Function and Optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
# Training the model
for epoch in range(300):
    y_pred = model(X) # forward propagation
    loss = criterion(y_pred, y) # compute the error between the predictions and true value
    
    optimizer.zero_grad() # set the gradients to 0, Pytorch automatically adds the gradients by default so this line is needed !!
    loss.backward() # compute loss gradients relatively to all the model parameters
    optimizer.step() # Adam optimizer uses the computed gradients to update the model weights
    
    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.4f}")
```

```python
print("Prediction example:")
print(model(torch.tensor([[1.0]])))
```
